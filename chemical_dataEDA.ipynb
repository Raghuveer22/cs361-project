```python
data=pd.read_csv('wqi.csv')
```

```python
# Define the columns for which you want to detect outliers
columns_of_interest = ['Ammonia-Total (as N)', 'True Colour']

# Calculate the first and third quartiles (Q1 and Q3) for each column
Q1 = data[columns_of_interest].quantile(0.25)
Q3 = data[columns_of_interest].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the threshold for outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Find the rows containing outliers for column1 and column3
outliers_column1 = data[(data['Ammonia-Total (as N)'] < (Q1['Ammonia-Total (as N)'] - threshold * IQR['Ammonia-Total (as N)'])) | (data['Ammonia-Total (as N)'] > (Q3['Ammonia-Total (as N)'] + threshold * IQR['Ammonia-Total (as N)']))]
outliers_column3 = data[(data['True Colour'] < (Q1['True Colour'] - threshold * IQR['True Colour'])) | (data['True Colour'] > (Q3['True Colour'] + threshold * IQR['True Colour']))]


outliers_column3.describe()
```
